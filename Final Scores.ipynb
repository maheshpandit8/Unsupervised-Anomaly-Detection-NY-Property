```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
%%time
data = pd.read_csv('Data/NY property data.csv')
```

```python
data.head()
```

```python
%%time
pca_data = pd.read_csv('Data/pca_output.csv')
```

```python
pca_data = pca_data.sort_values(by = 'pca_score', ascending = False)
pca_data.head()
```

```python
pca_data = pca_data.reset_index().drop(columns = ['index']).reset_index().rename(columns = {'index': 'pca_rank'})
```

```python
pca_data.head()
```

```python
%%time
autoencoder_data = pd.read_csv('Data/autoencoder_output.csv')
```

```python
autoencoder_data = autoencoder_data.sort_values(by = 'autoencoder_score', ascending = False)
autoencoder_data.head()
```

```python
autoencoder_data = autoencoder_data.reset_index().drop(columns = ['index']).reset_index().rename(columns = {'index': 'autoencoder_rank'})
```

```python
autoencoder_data.head()
```

```python
%%time
data = data.merge(pca_data, on = 'RECORD', how = 'left')
```

```python
%%time
data = data.merge(autoencoder_data, on = 'RECORD', how = 'left')
```

```python
data.head()
```

```python
%%time
data.to_csv('Data/NY data with scores.csv', index = False)
```

```python
data['rank'] = (data['pca_rank'] + data['autoencoder_rank']) / 2
```

```python
data.sort_values(by = 'rank', ascending = True).head()
```

```python
%%time
data.to_csv('Data/NY data with ranks.csv', index = False)
```
